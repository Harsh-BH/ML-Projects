```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

```

```python
df.columns
```

```python
x = df.drop(['Email No.', 'Prediction'],axis =1)
y = df['Prediction']
```

```python

scaler1 = StandardScaler()
scaler2 = MinMaxScaler()
x_scaled1 = scaler1.fit_transform(x)
x_scaled2 = scaler2.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled1,y,test_size=0.2,random_state=42)

```

```python
model = LogisticRegression()

model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
score= accuracy_score(pred,y_test)
```

```python
score
```

Neural Network

```python
import tensorflow as tf


from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam


```

```python
model1 = Sequential([
  Dense(units=1600,activation='relu',input_shape=(x_scaled1.shape[1],)),
  Dense(units=800,activation='relu'),
   Dense(units=400,activation='relu'),
  Dense(units=1)
])

model1.compile(loss = BinaryCrossentropy(from_logits=True),optimizer=Adam(),metrics=['accuracy'])

model1.fit(x_train,y_train,epochs=10)

loss, accuracy = model1.evaluate(x_test, y_test)
print(f'Test Accuracy: {accuracy}')
```

```python
import numpy as np

pred1 = model1.predict(x_test)



prob = tf.nn.sigmoid(pred1).numpy()
print("Predicted probabilities on x_test:")
print(prob)

for i in range(prob.shape[0]):
    if prob[i] >= 0.5:
        print(f"Sample {i}: Predicted class: 1")
    else:
        print(f"Sample {i}: Predicted class: 0")
```
