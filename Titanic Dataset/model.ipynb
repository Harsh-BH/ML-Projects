```python
def safe_extract(ticket):
    if (ticket)=='LINE':
        return 1
    parts = ticket.split(' ')
    if len(parts)==2:
        return int(parts[1])
    if len(parts)==3:
        return int(parts[2])
    
    
    try:
        return float(ticket)
    except:
        return None   

df1['tkt']=df1['Ticket'].apply(safe_extract)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

import xgboost as xgb

from xgboost import XGBClassifier


from sklearn.ensemble import RandomForestClassifier
scaler = StandardScaler()
x = df2.drop(['Ticket','Embarked','Survived'],axis=1)
y=df2.Survived

x_scaled = scaler.fit_transform(x)





```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.1,random_state=42)
```

```python
model = LogisticRegression()
model1=RandomForestClassifier(n_estimators=500,random_state=42)
model2=XGBClassifier(objective='binary:logistic', n_estimators=100,random_state=42)

model2.fit(x_train,y_train)
```

```python
pred1= model2.predict(x_test)

score=accuracy_score(pred1,y_test)
score
```
